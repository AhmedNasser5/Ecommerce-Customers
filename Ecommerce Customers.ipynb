# **Data Pre-processing**

# **Data Spliting**

```python
from sklearn.model_selection import train_test_split
```

```python
ecom.head()
```

```python
x=ecom[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y=ecom['Yearly Amount Spent']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# **Model Bulding & Training**

# **Model Training**

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

# **Model Prediction**

```python
model.predict([[80,70,60,50]])
```

```python
model_predict=model.predict(x_test)
```

```python
model_predict
```

```python
y_test.values
```

# **Model Evaluation**

```python
model.score(x,y)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,accuracy_score
```

```python
MSE=mean_absolute_error(y_test,model_predict)
MSE
```

```python
r2_score = r2_score(y_test, model_predict)
r2_score
```
